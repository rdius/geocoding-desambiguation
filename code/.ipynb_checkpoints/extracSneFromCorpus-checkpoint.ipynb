```python
# nlp_en = spacy.load('en_core_web_sm')# Text with nlp
nlp_enL = spacy.load('en_core_web_trf')
```

```python
""" 
for each SNE extracted from the document, return its potential candidate
"""
def getCandidate_(listOfSne,lang):
    
    AllSneCandidateD = {}
    NotAsSne = {}
    for sne in listOfSne:
#         AllSneCandidateD['source']
        geo = geocoder.geonames(sne, key=key, maxRows=1000,featureClass=['A','P'], lang =lang) #,country=['GB']
        Candidates = list(set([(r.address, r.lat, r.lng, r.country_code ,r.feature_class, r.population) for r in geo]))
#         print("list or tuple ? :", Candidates)
        """ 
        Parmi les candidats retournés, ne garder que ceux dont les noms correspondent 
        vraiment au nom de la sne
        """
        SameNameAs = [tuple(sn) for sn in Candidates if sn[0]==sne]
        AllSneCandidateD[sne] = SameNameAs
        """
        Traiter le cas des SNEs dont les noms ne matchent pas exactement avec les candidats trouvés
        """
        if len(AllSneCandidateD[sne]) == 0:
            print("zero len :", sne)
#             NotAsSne[sne] = Candidates
            Candidates = [tuple(sn) for sn in Candidates]
            NotAsSne[sne] = Candidates
    return AllSneCandidateD, NotAsSne
```
